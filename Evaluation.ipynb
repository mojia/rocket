```python
import pandas as pd

%run Evaluation_Metric.ipynb

class Evaluation:
    
    def __init__(self,account,context,policy_name):
        self.account = account
        self.context = context
        self.policy_name = policy_name
        self.policy_loader = Policy_Loader(self.policy_name)
        self.debug = self.policy_loader.enabled_log_debug()
        
        self.evaluation_metric = Evaluation_Metric(account)
        self.report = pd.DataFrame(columns=evaluation_columns)
        
        
    def get_evaluation_report_of_multip_stocks(self,start_date,end_date):
        return self.get_evaluation_report('all_symbols',start_date,end_date,pd.DataFrame(columns=operation_columns))
        
        
    def get_evaluation_report(self,symbol,stock_start_date,stock_end_date,
                             operation_df):
        record = [symbol,stock_start_date,stock_end_date,
                  self.policy_name,self.context['short'],self.context['mid']]
        
        in_hand_asset_drawdown = self.evaluation_metric.get_in_hand_asset_drawdown()
        record.append(in_hand_asset_drawdown)
        
        avg_drawdown = self.evaluation_metric.avg_max_drawdown()
        record.append(avg_drawdown)
        
        profits = operation_df['profit_rate'].values
        loss_times = self.evaluation_metric.continue_loss_times(profits)
        record.append(loss_times)
        
        float_profit_drawdown = self.evaluation_metric.get_float_profit_drawdown()
        record.append(float_profit_drawdown)
        
        avg_year_return = self.evaluation_metric.year_avg_return_rate()
        record.append(avg_year_return)
        
        year_regular_return = self.evaluation_metric.year_regular_return_rate()
        record.append(year_regular_return)
        
        year_returns = self.evaluation_metric.year_return_rate_array()
        cagrs = self.evaluation_metric.get_CAGR_array()
        if(self.debug==True):
            print('\nyear_returns\n'+str(year_returns))
            print('Compound avg year rate\n'+str(cagrs))
        
        mean_year_return = self.evaluation_metric.mean_year_return_rate()
        record.append(mean_year_return)
        
        std_year_return = self.evaluation_metric.std_year_return_rate()
        record.append(std_year_return)
        
        mean_month_return = self.evaluation_metric.mean_month_return_rate()
        record.append(mean_month_return)
        
        std_month_return = self.evaluation_metric.std_month_return_rate()
        record.append(std_month_return)
        
        profit_month = self.evaluation_metric.profit_month_rate()
        record.append(profit_month)
        
        win_rate_value = self.evaluation_metric.win_rate(profits)
        record.append(win_rate_value)
        
        mar_value = self.evaluation_metric.mar()
        record.append(mar_value)
        
        sharp_value = self.evaluation_metric.sharp()
        record.append(sharp_value)
        
        rar = self.evaluation_metric.get_RAR()
        record.append(rar)
        
        record.append(operation_df.shape[0])

        self.report.loc[self.report.shape[0]+1] = record
        
        return self.report
        
        
        
```
