```python
import datetime as dt
from datetime import datetime
import time
import os


def tostr(float_number):
    return '%.2f' % float_number

def out_of_range(date,start_datetime,end_datetime):
    year_s, mon_s, day_s = date.split('-')
    cur_datetime = datetime(int(year_s), int(mon_s), int(day_s))
    
    return cur_datetime<start_datetime or cur_datetime>end_datetime
    
```

```python
def get_all_symbols(home_dir):
    symbol_list = []
    for line in open(home_dir+'symbols.txt'):
        line = line.strip()
        items = line.split(',')

        symbol_list += items

    return symbol_list


def get_symbols(scale,home_dir):
    symbols = get_all_symbols(home_dir)
    return symbols[:scale]


def get_symbol_slices(start_index,end_index,home_dir):
    symbols = get_all_symbols(home_dir)
    return symbols[start_index:end_index]
```
